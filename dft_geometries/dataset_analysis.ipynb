```python
#histogram distribution of number of atoms
import pandas as pd
import seaborn as sns

df = pd.read_csv("/home/aweiland/comparing_theories/5k_mol_stats.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="darkgrid")

fig = plt.figure(figsize=(12, 8))
gs = fig.add_gridspec(2, 2, height_ratios=[1, 1])

ax1 = fig.add_subplot(gs[0, :])
sns.histplot(data=df, x="num_atoms", binwidth = 3, color ="orange", ax=ax1)
ax1.set_title("Distribution of Total Atoms")
ax1.set_xlabel("Total Atoms")
ax1.set_ylabel("Count")
ax1.set_xticks(np.arange(0, 175, 10))


plt.show()
```
