```python
import os

for dirpath, dirnames, files in os.walk("/home/aweiland/comparing_theories/xTB/final_run/outputs"):
    for file in files:
        if file.endswith(".out"):
            
            try:
                mol_id = int(file.strip("xtb_").strip(".out"))
            except:
                print(file)
                continue
            file_path = os.path.join(dirpath, file)
            
            f = open(file_path, "r")
            line = next(f)


            while line is not None:
                if "Sum of individual times" in line:
                    time = float(line.split()[5])
                    df.loc[df["mol_id"] == mol_id, "xtb_runtime"] = time

                    f.close()
                    break
                line = next(f, None)
            
    




```

```python
import pandas as pd
df = pd.read_csv("/home/aweiland/comparing_theories/5k_mol_stats.csv")
df["uma_time"] = None
df.head()
```

```python
#getting times for uma
#Script to get the time taken for all egret calculations and put them
#into a dataframe



import os
from datetime import datetime

batch_dir_path = "/home/aweiland/comparing_theories/UMA/final_uma/outputs"



for sub_dir in os.listdir(batch_dir_path):

    mol_id = int(sub_dir.strip("mol_"))

    sub_dir_path = os.path.join(batch_dir_path, sub_dir)

    log_dir = os.path.join(sub_dir_path, "opt.log")


    with open(log_dir, "r") as f:
        lines = f.readlines()
        try:
            start = lines[1].split()[2]
            end = lines[-1].split()[2]
        except IndexError:
            pass



        fmt = "%H:%M:%S"
        t1 = datetime.strptime(start, fmt)
        t2 = datetime.strptime(end, fmt)

        #handles situation where job starts one day and ends another 
        #which could create a negative time
        if start[0] == "2" and end[0] == "0":
            start = "1-" + start
            end = "2-" + end
            fmt = "%d-%H:%M:%S"
            t1 = datetime.strptime(start, fmt)
            t2 = datetime.strptime(end, fmt)


        delta = t2 - t1

        seconds = delta.total_seconds()

        df.loc[df["mol_id"] == mol_id, "uma_time"] = seconds

df.head()

```

```python
df.to_csv("/home/aweiland/comparing_theories/5k_mol_stats.csv", index=False)
```
