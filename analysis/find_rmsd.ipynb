```python
# calculate rmsd for all xtb molecules
import pandas as pd 
df = pd.DataFrame()
df["mol_id"] = None
df["num_atoms"] = None
df["xtb_rmsd"] = None
df["aimnet_rmsd"] = None


```

```python
from MDAnalysis.analysis.rms import rmsd
import MDAnalysis as mda
import os


xtb_directory = "/home/aweiland/comparing_theories/xTB/final_run/all_xyz"
dft_directory = "/home/aweiland/comparing_theories/dft_geometries/dft_xyz_files"

for file in os.listdir(xtb_directory):
    mol_id = file.strip("xtb_").strip(".xyz")
    
    xtb_path = os.path.join(xtb_directory, file)
    
    dft_file = "dft_" + mol_id + ".xyz"
    dft_path = os.path.join(dft_directory, dft_file)
    
    dft_mol = mda.Universe(xtb_path)
    xtb_mol = mda.Universe(dft_path)

    dft_atoms = dft_mol.atoms
    xtb_atoms = xtb_mol.atoms
    
    value = rmsd(dft_atoms.positions, xtb_atoms.positions, center=True, superposition=True)
    
    df.loc[len(df)] = [mol_id, len(dft_atoms), value, None]
    
    
df.head()
    
```

```python
from MDAnalysis.analysis.rms import rmsd
import MDAnalysis as mda
import os


xtb_directory = "/home/aweiland/comparing_theories/aimnet/aimnet_xyz_files"
dft_directory = "/home/aweiland/comparing_theories/dft_geometries/dft_xyz_files"

for file in os.listdir(xtb_directory):
    mol_id = file.strip("aimnet_").strip(".xyz")
    
    xtb_path = os.path.join(xtb_directory, file)
    
    dft_file = "dft_" + mol_id + ".xyz"
    dft_path = os.path.join(dft_directory, dft_file)
    
    dft_mol = mda.Universe(xtb_path)
    xtb_mol = mda.Universe(dft_path)

    dft_atoms = dft_mol.atoms
    xtb_atoms = xtb_mol.atoms
    
    value = rmsd(dft_atoms.positions, xtb_atoms.positions, center=True, superposition=True)
    
    if (df['mol_id'] == mol_id).any():
        df.loc[df['mol_id'] == mol_id, 'aimnet_rmsd'] = value
    
    else:
        df.loc[len(df)] = [mol_id, len(dft_atoms), None, value]

    
    
df.head()
```

```python
df.to_csv("/home/aweiland/comparing_theories/mol_stats.csv")
```
