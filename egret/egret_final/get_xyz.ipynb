```python
import os

for i in range(15):

    batch_path = f"/home/aweiland/comparing_theories/egret/egret_final/output/batch_{i}"

    all_xyz_path = "/home/aweiland/comparing_theories/egret/egret_final/egret_xyz"

    for directory in os.listdir(batch_path):
            subdir = os.path.join(batch_path, directory)
            file = os.path.join(subdir, "opt.traj")
            try:
                traj = Trajectory(file)
            except:
                print(directory)
                continue
            atoms = traj[-1]


            positions = atoms.get_positions()  # shape: (n_atoms, 3)
            final_positions = []

            for row in positions:
                fixed_row = [float(coord) for coord in row]  # optional: already floats
                final_positions.append(fixed_row)


            symbols = atoms.get_chemical_symbols()

            mol_id = directory.strip("mol_")

            output_file = os.path.join(all_xyz_path, "egret" + mol_id + ".xyz")
            f = open(output_file, "w")
            f.write(str(len(symbols))+ "\n\n")

            for i in range(len(symbols)):
                f.write(symbols[i] + "   " + str(final_positions[i][0]) 
                        + "   " + str(final_positions[i][1]) + "   " 
                        + str(final_positions[i][2]) + "\n")
            f.close()


```

```python
#get time

import os
from datetime import datetime

for i in range(15):

    batch_path = f"/home/aweiland/comparing_theories/egret/egret_final/output/batch_{i}"

    all_xyz_path = "/home/aweiland/comparing_theories/egret/egret_final/egret_xyz"

    for directory in os.listdir(batch_path):
            subdir = os.path.join(batch_path, directory)
            file = os.path.join(subdir, "opt.log")
            
            mol_id = int(directory.strip("mol_"))
            
            with open(file, "r") as f:
                lines = f.readlines()
                try:
                    start = lines[1].split()[2]
                    end = lines[-1].split()[2]
                except IndexError:
                    pass



                fmt = "%H:%M:%S"
                t1 = datetime.strptime(start, fmt)
                t2 = datetime.strptime(end, fmt)

                #handles situation where job starts one day and ends another 
                #which could create a negative time
                if start[0] == "2" and end[0] == "0":
                    start = "1-" + start
                    end = "2-" + end
                    fmt = "%d-%H:%M:%S"
                    t1 = datetime.strptime(start, fmt)
                    t2 = datetime.strptime(end, fmt)


                delta = t2 - t1

                seconds = delta.total_seconds()

                df.loc[df["mol_id"] == mol_id, "egret_runtime"] = seconds





```
