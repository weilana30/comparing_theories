```python
import os

input_dir = "/home/aweiland/comparing_theories/xTB/input_files"
output_root = "/home/aweiland/comparing_theories/xTB/final_run/outputs"

not_found = []
i = 0

# ðŸ§  Step 1: Collect all subdirectory names under outputs/
output_subdirs = set()
for dirpath, dirnames, _ in os.walk(output_root):
    for d in dirnames:
        output_subdirs.add(d)

# ðŸ§  Step 2: Loop through all .inp files in input_files/
for filename in os.listdir(input_dir):
    if filename.endswith(".inp") and filename.startswith("xtb_"):
        i += 1
        if i % 500 == 0:
            print(f"Checked {i} input files...")

        base_name = os.path.splitext(filename)[0]  # Remove ".inp"

        if base_name not in output_subdirs:
            not_found.append(filename)
            print("No matching output subdirectory found for:")
            print(filename)
            print()

print(f"\nâœ… Total input files without a corresponding output directory: {len(not_found)}")
```
